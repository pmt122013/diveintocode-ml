# Going down Mt. Fuji
Gradient descent is likened to going down a mountain or valley. You will learn the gradient descent method by creating a descent program from Mt. Fuji, as **if it were an objective function that you want** to minimize.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Load data
In order from the left, "the number of that point", "latitude", "longitude", "elevation", "distance from point 0". The units of altitude and distance are meters. Data for 300 locations.

```python
csv_path = 'data/mtfuji_data.csv'

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
print(fuji[130:140])
```

## [Problem 1] Visualization of data

```python
plt.title('Relation between position and elevation of Mt.Fuji')
plt.xlabel('position')
plt.ylabel('elevation [m]')
plt.plot(fuji.T[0], fuji.T[3])
plt.show()
```

## [Problem 2] Create a function to calculate the slope of a point
Create a function that calculates the slope based on the relationship between the "current point number" and the "current point number (the current point number-1)".

The slope was calculated by the following formula. In this data, the "elevation" corresponds to y and the "point number" corresponds to x.

```python
def find_gradient(x, y):
    amount_x = x[1:] - x[:-1]
    amount_y = y[1:] - y[:-1]
    gradient = amount_y / amount_x
    return gradient

slope = find_gradient(fuji.T[0], fuji.T[3])
plt.xlabel('position')
plt.ylabel('elevation [m]')
plt.plot(fuji.T[0], fuji.T[3], label='linear')
plt.plot(fuji.T[0][1:], slope, label='gradient')
plt.legend()
plt.show()
```

## [Problem 3] Create a function to calculate the destination
Create a function that calculates the next destination based on the information on the slope of the current location. Expressed as a formula:

$\text{Destination point} = \text{current location} - \alpha × \text{Gradient at the current location}$

$\alpha$ is the **hyper-parameters** to take any value greater than 0, called **learning rate**. The way of descending changes depending on this value. Here, for example, Settings it to 0.2.

```python
def find_destination(cur_location, fuji_data):
    alpha = 0.2
    point = fuji_data.T[0]
    elevation = fuji_data.T[3]
    gradient = find_gradient(point[cur_location-1:cur_location+1], elevation[cur_location-1:cur_location+1])
    
    destination = cur_location - alpha * gradient
    if destination <= 0:
        return 0
    else:
        return round(destination[0])
```

```python
find_destination(136, fuji)
```

```python
cur_location = 136
path = []
while cur_location > 0:
    des = find_destination(cur_location, fuji)
    path.append(des)
    cur_location = des
print(path)
```
