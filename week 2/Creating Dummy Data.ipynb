# Creating dummy data
When experimenting with programs related to machine learning, there are times when you want to have random data (dummy data) that is not real data. Let's get familiar with NumPy by creating dummy data with random numbers that follow a normal distribution.

### Creating random number

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
mean1 = [-3, 0]
mean2 = [0, -3]
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

x1, y1 = np.random.multivariate_normal(mean1, cov, 500).T
x2, y2 = np.random.multivariate_normal(mean2, cov, 500).T
```

### Visualization

* Using plot

```python
plt.title('Normal plot of (x1, y1) and (x2, y2)')
plt.plot(x1, y1, 'x', color='orange', label='0')
plt.plot(x2, y2, 'x', color='green', label='1')
plt.legend()
plt.show()

plt.title('Scatter plot of (x1, y1) and (x2, y2)')
plt.scatter(x1, y1, color='purple', marker='o', label='0')
plt.scatter(x2, y2, color='cyan', marker='o', label='1')
plt.legend()
plt.show()

plt.title('Histogram plot of (x1, y1)')
plt.hist(x1, color='#2ecc71', label='x')
plt.hist(y1, color='#e67e22', label='y')
plt.xlim(-20, 20)
plt.legend()
plt.show()

plt.title('Histogram plot of (x2, y2)')
plt.hist(x2, color='#2ecc71', label='x')
plt.hist(y2, color='#e67e22', label='y')
plt.xlim(-20, 20)
plt.legend()
plt.show()
```

### Data combination

```python
new_x = np.concatenate((x1, x2))
new_y = np.concatenate((y1, y2))
```

### Visualization of combined data

```python
plt.title('Normal plot')
plt.plot(new_x, new_y, 'x', color='orange', label='combined')
plt.legend()
plt.show()

plt.title('Scatter plot')
plt.scatter(new_x, new_y, color='purple', marker='o', label='combined')
plt.legend()
plt.show()

plt.title('Histogram plot of x')
plt.hist(new_x, color='#2ecc71', label='x')
plt.xlim(-20, 20)
plt.legend()
plt.show()

plt.title('Histogram plot of y')
plt.hist(new_y, color='#e67e22', label='y')
plt.xlim(-20, 20)
plt.legend()
plt.show()
```
