# Sorori Shinzaemon Problem
When Hideyoshi offered him a reward, Shinzaemon was asked what he would like to receive, and he asked for one grain of rice today, two grains the next day, four grains the day after that, and so on, doubling the amount of rice each day for 100 days. Thinking that a grain of rice would be no big deal, Hideyoshi readily agreed, but realized halfway through that doubling the amount of rice each day would result in a huge amount of rice after 100 days, so he had the reward changed to something else.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Calculation

```python
def compute_sorori_shinzaemon(day):
    grain_amount_lst = [1]
    total_lst = [1]
    
    for i in range(1, day):
        grain_amount_lst.append(grain_amount_lst[i-1]*2)
        total_lst.append(total_lst[i-1] + grain_amount_lst[i])
    return grain_amount_lst, total_lst

grain_amount_lst, total_lst = compute_sorori_shinzaemon(100)
```

## Visualization

```python
%matplotlib inline
# Grain on each day
plt.title('Amount of grain after 100 days')
plt.xlabel('Days')
plt.ylabel('Number of rice')
plt.plot(grain_amount_lst, marker='o', mfc='yellow', markersize='7')
plt.show()

# Total grain on each day
plt.title('Total amount of grain after 100 days')
plt.xlabel('Days')
plt.ylabel('Number of rice')
plt.plot(total_lst, color='orange',marker='o', mfc='yellow', markersize='7')
plt.show()
```

### How many people can live for how many days with the rice we get?

```python
RICE_PER_DAY = 19200
def days_to_live(grain_numb, people_numb):
    return grain_numb / (people_numb * RICE_PER_DAY)

people_numb = 10000000000000000
res = days_to_live(total_lst[-1], people_numb)
print('With {} people, they can live for {} days'.format(people_numb, res))
```
