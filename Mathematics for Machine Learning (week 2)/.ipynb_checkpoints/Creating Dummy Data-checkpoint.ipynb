```python
import numpy as np
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

# Problem 1: Creating Random Numbers

```python
def creating_random_numbers(mean):
    
    matrix = [[1.0, 0.8], 
              [0.8, 1.0]]
    
    np.random.seed(0)
    ndarray = np.random.multivariate_normal(mean, matrix, 500)
    return ndarray
    
ndarray = creating_random_numbers(mean = [-3, 0])

print(ndarray)
```

# Problem 2: Visualization by scatter plot

```python
def scatter_plot_visualization(ndarray):
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Scatter")
    plt.scatter(ndarray[:, 0], ndarray[:, 1]) 
    plt.show()
    
scatter_plot_visualization(ndarray)
```

# Problem 3: Visualization by histogram

```python
def visualization_histogram_1():
    plt.xlabel("x1")
    plt.ylabel("frequency")
    plt.title("Histogram of X1")
    plt.xlim(left = -7, right = 2)
    plt.hist(ndarray[:, 0], bins = 50, density = 1) 

visualization_histogram_1()
```

```python
def visualization_histogram_2():
    plt.xlabel("x2")
    plt.ylabel("frequency")
    plt.title("Histogram of X2")
    plt.xlim(left = -5, right = 3)
    plt.hist(ndarray[:, 1], bins = 50, density = 1, color = "green") 
    
visualization_histogram_2()
    
```

# Problem 4: Addition of Data

```python
ndarray_1 = creating_random_numbers(mean = [0, -3])

def scatter_combine_addition_data(ndarray, ndarray_1):
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Scatter")
    plt.scatter(ndarray[:, 0], ndarray[:, 1], label = 0, color = "blue") 
    plt.scatter(ndarray[:, 1], ndarray[:, 0], label = 1, color = "orange")
    plt.legend()
 
plt.show()    
    
scatter_combine_addition_data(ndarray, ndarray_1)
```

# Problem 5: Data combination

```python
def data_combination(ndarray, ndarray_1):
    data1 = np.concatenate([ndarray, ndarray_1], axis = 0)
    data2 = np.vstack([ndarray, ndarray_1])

    return data1, data2

data1, data2 = data_combination(ndarray, ndarray_1)

print("Data 1 shape is: {}".format(data1.shape))
print("Data 2 shape is: {}".format(data2.shape))
```

# Problem 6: Labeling

```python
def data_labeling(data):
    data_zeros = np.zeros((500, 1))
    data_ones = np.ones((500, 1))

    data_zeros_ones = np.vstack([data_zeros, data_ones])

    label_data = np.concatenate([data, data_zeros_ones], axis = 1)
    
    return label_data

#np.full can replace np.ones and np.zeros by np.full((500, 1), value)

label_data = data_labeling(data1)

print("Label data shape is: {}".format(label_data.shape))
print("Label data: \n {}".format(label_data))
```
