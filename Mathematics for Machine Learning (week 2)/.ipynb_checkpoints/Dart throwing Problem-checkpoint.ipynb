```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

# Problem 1: Choosing a dart 

```python
dice_indexes = [1, 2, 3, 4, 5, 6]
dice_choice = np.random.choice(dice_indexes)

print("Dice roll: {}".format(dice_choice))
```

# Problem 2: Throw darts and record the place

```python
# Using np.random.normal()

x_loc = np.random.normal(loc = dice_choice, scale = 0.1)
y_loc = np.random.normal(loc = 0, scale = 0.1)

loc1 = np.array([x_loc, y_loc])

print("Location using np.random.normal(): \n {}".format(loc1))
```

```python
# Using np.random.multivariate_normal()

mean = [dice_choice, 0]
cov = [[0.01, 0], 
       [0, 0.01]]

loc2 = np.random.multivariate_normal(mean, cov)

print("Location using np.random.multivariate_normal(): \n {}".format(loc2))
```

# Problem 3:  Throw a lot of darts and record them in a state of being tied to the number

```python
n_darts = 1000
loc_darts_dict ={}

dice_indexes = [1, 2, 3, 4, 5, 6]

for i in range(n_darts):
    dice_choice = np.random.choice(dice_indexes)
    x_loc = np.random.normal(loc = dice_choice, scale = 0.1)
    y_loc = np.random.normal(loc = 0, scale = 0.1)
    if dice_choice in loc_darts_dict:
        loc_darts_dict[dice_choice] = np.concatenate((loc_darts_dict[dice_choice], [[x_loc, y_loc]]))
    else:
        loc_darts_dict[dice_choice] = np.array([[x_loc, y_loc]])

print("Example, All locations of dice choice {}: \n {}".format(2, loc_darts_dict.get(2)))        
```

# Problem 4: Drawing the position of the darts


```python
plt.figure(figsize = (20, 3))
ax = plt.axes()

for key in loc_darts_dict.keys():
    plt.scatter(loc_darts_dict[key][:, 0], loc_darts_dict[key][:, 1])
    circle1 = mpatches.Circle(xy = (key, 0), radius = 0.5, ec = "b", fill = False)
    circle2 = mpatches.Circle(xy = (key, 0), radius = 0.25, ec = "b", fill = False)
    ax.add_patch(circle1)
    ax.add_patch(circle2)
    
ax.set_title("n_darts = {}, n_boards = {}".format(1000, 6))    
ax.set_xlim(0, 7)
ax.set_ylim(-0.5 , 0.5)

plt.show()
```

# Problem 5: Functionalization

```python
def draw_darts(n_darts = 1000, n_boards = 6):
    darts_locs_dict = {}
    dice_new_indexes = list(range(1, n_boards + 1))
    
    for i in range(n_darts):
        dice_choice = np.random.choice(dice_new_indexes)
        x_loc = np.random.normal(loc = dice_choice, scale = 0.1)
        y_loc = np.random.normal(loc = 0, scale = 0.1)
        if dice_choice in darts_locs_dict:
            darts_locs_dict[dice_choice] = np.concatenate((darts_locs_dict[dice_choice], [[x_loc, y_loc]]))
        else:
            darts_locs_dict[dice_choice] = np.array([[x_loc, y_loc]])  
            
    plt.figure(figsize = (n_boards*3 + 2, 3))
    ax = plt.axes()

    for key in darts_locs_dict.keys():
        plt.scatter(darts_locs_dict[key][:, 0], darts_locs_dict[key][:, 1])
        circle1 = mpatches.Circle(xy = (key, 0), radius = 0.5, ec = "b", fill = False)
        circle2 = mpatches.Circle(xy = (key, 0), radius = 0.25, ec = "b", fill = False)
        ax.add_patch(circle1)
        ax.add_patch(circle2)
    ax.set_title("n_darts = {}, n_boards = {}".format(n_darts, n_boards))
    ax.set_xlim(0, n_boards + 1)
    ax.set_ylim(-0.6 , 0.6)

    plt.show() 

for i in range(1, 10):
    draw_darts(n_darts= i*200, n_boards = i)
    
    
```
