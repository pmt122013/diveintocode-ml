```python
import matplotlib.pyplot as plt
```

Problem 1:

```python
def rice_on_100_th_day(day = 100, n_first_day_rices = 1):
    list_n_grains = [n_first_day_rices]
    list_total_grains = [list_n_grains[0]]
    for i in range(day - 1):
        list_n_grains.append(list_n_grains[-1]*2)
        list_total_grains.append(list_total_grains[-1] + list_n_grains[-1])
        
    return list_n_grains,list_total_grains
day = 100

list_n_grains, list_total_grains = rice_on_100_th_day(day = day, n_first_day_rices = 1)

plt.title("Sorori Shinzaemon's rices")
plt.xlabel("days")
plt.ylabel("Number of rices received")
plt.plot(range(1, day+1), list_n_grains, label = "per day")
plt.plot(range(1, day+1), list_total_grains, label = "total")
plt.legend()
plt.show()

print("Number of rice in day 100th: {} grains".format(list_n_grains[-1]))
print("Total number of rice in day 100th: {} grains".format(list_total_grains[-1]))
```

Problem 2:

```python
def rice_on_n_th_day(day = 100, n_first_day_rices = 1):
    list_n_grains = [n_first_day_rices]
    list_total_grains = [list_n_grains[0]]
    for i in range(day - 1):
        list_n_grains.append(list_n_grains[-1]*2)
        list_total_grains.append(list_total_grains[-1] + list_n_grains[-1])
        
    return list_n_grains,list_total_grains
day = int(input("Enter your day: "))
n_first_day_rices = int(input("Enter your starting rice amount: "))
list_n_grains, list_total_grains = rice_on_n_th_day(day = day, n_first_day_rices = n_first_day_rices)

plt.title("Sorori Shinzaemon's rices")
plt.xlabel("days")
plt.ylabel("Number of rices received")
plt.plot(range(1, day+1), list_n_grains, label = "per day")
plt.plot(range(1, day+1), list_total_grains, label = "total")
plt.legend()
plt.show()

print("Number of rice in day {}: {} grains".format(day, list_n_grains[-1]))
print("Total number of rice in day {}: {} grains".format(day, list_total_grains[-1]))
```

Problem 3:
Calorie in 1 grain is 77.28 cal
A person needed 2200*100 calorie per day

```python
def compute_how_many_days(n_people, n_grains, calories_of_rice = 77.28, calories_consumption = 2200*1000):
    total_calories = n_grains * calories_of_rice
    day_to_live = total_calories//(calories_consumption*n_people)
    return day_to_live

day = int(input("Enter your day: "))
n_first_day_rices = int(input("Enter your starting rice amount: "))
n_people = int(input("Enter your total villagers: "))

list_n_grains, list_total_grains = rice_on_n_th_day(day = day, n_first_day_rices = n_first_day_rices)
n_grains = list_total_grains[-1]
days_to_live = compute_how_many_days(n_people,n_grains)

print("Total grains on day {}: {} grains".format(day, list_total_grains[-1]))
print("{} people can survive {} days".format(n_people, days_to_live))
```
